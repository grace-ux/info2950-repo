## Why was this dataset created?

### Korean Tourism Dataset: 
The dataset was created as a foundation for amplification of the Korean tourism market. It also lays the framework for formulating future marketing strategies, thus allowing Korea to expand and tailor the tourism experience towards more targeted individuals. 

The dataset identifies the total number of visitors that came into South Korea from 2016- 2023. It also contains the purpose of visit(tourism, business, education, etc.), mode of transportation, gender, and whether the visitors were flight attendants or not (to differentiate from “real” visitors to flight attendants). 

### Reddit Dataset:
The dataset was created to maintain a record of officially announced K-pop events worldwide from 2016-2023, such as concerts, fanmeets, and festivals, specifically for the /r/kpop subreddit community. It serves as an archive for reference and planning purposes for K-pop fans who use Reddit.

### Youtube Dataset
The dataset was created to organize the top trending videos on Youtube each day, as an updated version of the Youtube Trending Video Statistics and Comments dataset. It is a daily record of the top trending videos, which Youtube identifies through a combination of factors including a measure of user interactions (e.g. likes, views, comments).

## Who funded the creation of this dataset?

### Korean Tourism Dataset:
The Korean government has a separate entity called “Visit Korea Datalab” (by the Korean Tourism Organization) which focuses on the research of Korean tourism. They funded the creation of the dataset by surveying visitors to Korea and integrating pre-existing datasets from “UNWTO World Tourism Barometer.” 

The datasets can be used freely under Korean Copyright Law if they are provided by the Korean Tourism Organization, and there is no need to provide any special form of acknowledgement as long as a link to one of its homepages (kto.visitkorea.or.kr) is provided.

The link to the Datalab itself is: https://datalab.visitkorea.or.kr/datalab/portal/main/getMainForm.do

### Reddit Dataset:
There is no specific mention of funding; such initiatives on Reddit are usually community-driven and supported by volunteers. The creation and maintenance of this dataset are likely a collective effort by /r/kpop subreddit members rather than being externally funded.

### Youtube Dataset: 
This license for this dataset CC0: Public Domain and is publicly funded, with 54 total unique contributors. It was created using https://youtube.com’s API. Our raw data can be found at https://www.kaggle.com/datasets/rsrishav/youtube-trending-video-dataset/data.

## Exhibit 4: Total view count over time
A graph of the total view count on kpop Youtube videos over the time period of 2020-2023.

Date: 2020-2023

Intention: We hope to understand the larger spike in the data, what caused it, and if it has to do with our query on Youtube bots. We also want to discover patterns within the views of trending Youtube videos, and if those patterns correlate to any patterns in tourism over time.

```python
combined_df['publishedAt'] = pd.to_datetime(combined_df['publishedAt'])
country_views = \
combined_df.groupby('country')['view_count'].sum().reset_index()
top_countries = country_views.nlargest(5, 'view_count')
selected_rows = \
combined_df[combined_df['country'].isin(top_countries['country'])]
plt.figure(figsize=(12, 6))
sns.lineplot(data=selected_rows, \
             x='publishedAt', \
             y='view_count', hue='country', palette='bright')

plt.xlabel('Time Published')
plt.ylabel('Views')
plt.title('Kpop Youtube Views Over Time for Top 5 Countries')
plt.xticks(rotation=45)
plt.legend(title='Country', loc='upper left')

plt.show()
```

## Data Limitations

### Korean Tourism Dataset:
Coverage gaps exist, notably for some African countries, labeled as "etc" or "misc," likely due to low visitor numbers, which might introduce a focus bias. Data exclusions during cleanup could result in sampling bias, potentially affecting the representativeness of traveler profiles or travel purposes. The datasets' "purpose of visit" categories may lack specificity, posing a challenge in reflecting multiple travel intentions. Translation through tools like Google Spreadsheets may have led to the loss of cultural and linguistic subtleties, possibly affecting the accuracy of country names and purpose classifications. The datasets only recognize binary gender options, which may not align with all visitors' identities, highlighting the need for sensitive and accurate gender classification. Lastly, there's an inconsistency in processing, particularly with the Gulf Cooperation Council (GCC) countries' representation in the geo JSON file versus dataframes, resulting in the complete omission of these countries from the map.

### Reddit Dataset:
The dataset is catered more towards Reddit users, which may not reflect the broader K-pop fan base. The data collection is driven by community efforts, lacking formal funding, which might influence the completeness and systematic update of the records. The data is reliant on official confirmations, which means unofficial or lesser-known events may be excluded. This potentially overlooks significant segments of k-pop activity. The recording process is very selective, which omits events that don’t meet the subreddit’s criteria or failing to update canceled events promptly. It also focuses solely on events where artists are physically present, disregarding virtual engagements that have become increasingly relevant.

### Youtube Dataset:
The absence of China—a significant market where local platforms dominate—suggests potential future adjustments to account for such alternatives. Recent changes to YouTube's dislikes feature have led to inconsistent data within the dataset, affecting comparative analysis. The dataset's focus on trending videos from a select group of countries overlooks the global reach of YouTube, resulting in incomplete data. There's also a specific issue with Japanese groups being categorized differently based on their marketing or association with Korean entertainment, which could confuse data analysis. Additionally, inconsistencies in engagement metrics, like likes and comments compared to fluctuating view counts, may point to changes in YouTube's view tracking methods. Lastly, the presence of bots artificially inflating views of K-pop videos calls into question the reliability of the data.

### Hypothesis 2
#### Model Explanatory Power and Specification Issues 
The OLS regression model's low R-squared value of 0.336 reveals a substantial limitation, indicating that a large proportion of the variability in the number of visitors to South Korea remains unexplained by the model. This suggests that additional factors influencing tourism have not been accounted for, pointing to the possibility of omitted variable bias. Furthermore, the negative constant coefficient is a red flag, suggesting a potential mis-specification of the model, as it indicates a negative number of visitors when all predictor variables are at zero, which is not feasible in a real-world context. The insignificance of the 'Event Count' coefficient raises questions about the direct impact of K-pop events on tourism within the constraints of this model, hinting that the model may not be capturing the complex dynamics of cultural influence on travel behavior.

To address the low explanatory power indicated by the R-squared value, we plan to incorporate additional relevant variables into the model. We should consider economic indicators, demographic factors, or other cultural trends that could influence tourism. We will conduct a more detailed analysis involving age, income levels, and specific reasons for travel to unveil hidden patterns and relationships. For the negative constant coefficient, we should revisit the data for accuracy and consider a model re-specification. We might need to transform or scale variables to better fit the model. Additionally, introducing interaction terms to capture the combined effect of variables on the dependent variable could provide a more realistic representation than considering their effects in isolation.

#### Challenges with Multicollinearity and Insignificant Variables 
Multicollinearity, as suggested by the moderate condition number, also poses a problem, potentially leading to inflated variances of the estimated coefficients and making them unreliable. The lack of statistical significance in the 'Event Count' variable further complicates the interpretation, as it implies that the presence of K-pop events doesn't significantly affect the tourism numbers, based on the current model structure. This could either reflect reality or indicate that the model is missing key variables that could provide a more comprehensive understanding of what drives tourism to South Korea.

To address the issue of multicollinearity and identify the variables contributing to the problem, we plan to calculate variance inflation factors (VIFs) for each predictor variable. By doing so, we can determine if any variables have high VIFs and may need to be removed or combined with others to reduce redundancy. Additionally, for insignificant variables like 'Event Count', we should consider exploring different model structures or non-linear relationships that may better capture the dynamics between K-pop events and tourism. It would also be beneficial to examine different subsets of the data or segment the data based on certain characteristics to uncover more significant effects.

#### Statistical Assumptions and Need for Model Refinement 
Finally, the regression analysis appears to be further limited by potential serial correlation and non-normal residuals, both of which violate the assumptions of OLS and can lead to biased inference. Without checks for the normality of residuals and adjustments for any serial correlation, the model's predictive power and the reliability of its hypothesis tests may be compromised. These statistical challenges, combined with the inability of the model to establish causal relationships, suggest that further refinement is needed, possibly through the inclusion of more relevant variables or by employing different econometric techniques better suited for causal inference.
